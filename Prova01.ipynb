#**6.** Um determinado sistema linear e invariante no tempo possui resposta ao impulso $h[n]$ dada por:
$$h[n] = 0.8^nu[n] + (−0.5)^nu[n]$$


####**(a)** Plote a resposta ao impulso do sistema para $−10 ≤ n ≤ 40$.

```python
import numpy as np
import matplotlib.pyplot as plt

Nmin = -10
Nmax = 40

n = np.arange(Nmin,Nmax+1)
u = 1*(n >= 0)

h = 0.8**n*u + (-0.5)**n*u

plt.figure(figsize=(10,5))
plt.stem(n,h)
plt.xlabel('$n$')
plt.title('$h[n]$')

plt.tight_layout()
plt.show()
```

Entrada: $x(t) = e^{-2t}u(t)$

$$\frac{d^2y(t)}{dt^2}+ 4\frac{dy(t)}{dt} + 3y(t) = \frac{dx(t)}{dt} + 5x(t)$$
É a mesma coisa de:

$$y[n] = \frac{1}{2}\left(y[n − 1] + \frac{x[n]}{y[n − 1]}\right)$$

em que $x[n] = Au[n]$, sendo $A$ o valor no qual deseja-se estimar a raiz quadrada e $y[−1]$ é
uma estimativa inicial para a raiz quadrada de $A$. O algoritmo é resolvido de forma recursiva
para $n ≥ 0$. Após uma quantidade satisfatórias de iterações, $y[n]$ tenderá ao valor real de √$A$. Implemente esse algoritmo e verifique seus resultados para os seguintes casos, em
que $N$ é a quatidade de iterações realizadas pelo algoritmo. Como estimativa inicial, utilize
$y[−1] = A/2$.

```python
import numpy as np
import matplotlib.pyplot as plt

A = 5
N = 10

y1 = A/2
xn = 0
y = 0

yn = np.zeros(N)
for index in np.arange(0,N):
  xn = A*(index>=0)
  y = 1/2*(y1+(xn/y1))
  y1 = y
  yn[index] = y

raiz = yn[-1]
erro = np.abs((np.sqrt(A) - raiz))/np.sqrt(A)*100
print("raiz de", A ,"= ",raiz)
print("erro =", erro,"%")

n = np.arange(0,N)

plt.figure(figsize=(10,5))
plt.stem(n,yn)
plt.xlabel('n')
plt.title('$\sqrt{5}$')

plt.tight_layout()
plt.show()
```

#**13.** Sendo $x[n]$ o sinal de áudio original e $y[n]$ o sinal de áudio processado, implemente os seguintes efeitos de áudio:


####**(a)** Echo:
$$y[n] = x[n] + ηy[n − N]$$
em que $η$ é a profundidade do efeito $(0 < η < 1)$ e $N$ é a quantidade de amostras de
atraso, calculada como:
$$N = Fs(∆t)$$
tal que $F_s$ é a frequência de amostragem do sinal de áudio e $∆t$ é o atraso desejado (em
segundos).


```python
import numpy as np
import matplotlib.pyplot as plt
import IPython
from scipy.io import wavfile

Fs, x = wavfile.read('acoustic.wav')
x = x/32678.0
N = len(x)

time = np.arange(0,N)*(1/Fs)  #[0 1/Fs 2/Fs 3/Fs .... (N-1)/Fs]

plt.figure(figsize=(10,6))
plt.plot(time,x)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=x, rate=Fs)
```

Implementando o Echo:

$$y[n] = x[n] + ηy[n − N]$$

```python
echo = np.zeros(N)
delay = 2
Nd = int(delay*Fs)
n = 0.9999
for index in np.arange(0,N):
  if index - Nd < 0:
    echo[index] = x[index]
  else:
    echo[index] = x[index] + n*x[index - N]

plt.figure(figsize=(10,6))
plt.plot(time,echo)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=echo, rate=Fs)
```

####**(b)** Overdrive:

$$
y[n] =\begin{cases}
2x[n], & -1/3 \leq x[n] < 1/3 \\
-\frac{3-(2-3|x[n]|)^2}{3}, & -2/3 \leq x[n] < -1/3 \\
\frac{3-(2-3x[n])^2}{3}, & 1/3 \leq x[n] < 2/3 \\
1, & x[n] \geq 2/3 \\
-1, & x[n] < -2/3
\end{cases}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import IPython
from scipy.io import wavfile

Fs, x = wavfile.read('acoustic.wav')
x = x/32678.0
N = len(x)

time = np.arange(0,N)*(1/Fs)  #[0 1/Fs 2/Fs 3/Fs .... (N-1)/Fs]

plt.figure(figsize=(10,6))
plt.plot(time,x)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=x, rate=Fs)
```

Implementando o Overdrive:

```python
yover = np.zeros(N)
for index in np.arange(0,N):
  if x[index] >= -1/3 and x[index] <= 1/3:
    yover[index] = 2*x[index]
  elif x[index] >= -2/3 and x[index] <= -1/3:
    yover[index] = -(3-(2-3*abs(x[index]))**2)/3   #foi adicionado o "asb"
  elif x[index] >= 1/3 and x[index] <= 2/3:
    yover[index] = (3-(2-3*x[index])**2)/3
  elif x[index] >= 2/3:
    yover[index] = 1
  else:
    yover[index] = -1

plt.figure(figsize=(10,6))
plt.plot(time,yover)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=yover, rate=Fs)
```

####**(c)** Fuzz:
$$ye[n] = \frac{x[n]}{|x[n]|}\left[1 − \exp \left(\frac{ax^2[n]}{|x[n]|}\right)\right] \\
y[n] = ηye[n]+ (1 − η)x[n]$$
em que $a$ é o ganho (teste alguns valores próximos à 10) e $η$ é a profundidade do efeito
$(0 < η < 1)$.

Verifique os resultados dos efeitos de forma visual (plote o gráfico no tempo) e de forma
auditiva. Os efeitos deverão ser testados em cima do áudio *acoustic.wav*.

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython
from scipy.io import wavfile

Fs, x = wavfile.read('acoustic.wav')
x = x/32678.0
N = len(x)

time = np.arange(0,N)*(1/Fs)  #[0 1/Fs 2/Fs 3/Fs .... (N-1)/Fs]

plt.figure(figsize=(10,6))
plt.plot(time,x)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=x, rate=Fs)
```

Implementando o Fuzz:

$$
\frac{x[n]}{|x[n]|} = \begin{cases}
  -1 & x[n] \lt 0 \\
  1 & x[n] \gt 0
\end{cases}
$$

```python
yfuzz = np.zeros(N)
ye = 0
a = -3
d_deph = 0.1
for index in np.arange(0,N):
  if x[index] > 0:
    ye = (1 - np.exp(a*x[index]))
  else:
    ye = -(1 - np.exp(a*abs(x[index])))
  yfuzz[index] = d_deph*ye + (1 - d_deph)*x[index]

plt.figure(figsize=(10,6))
plt.plot(time,yfuzz)
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(data=yfuzz, rate=Fs)
```
