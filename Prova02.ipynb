**4)** Considere um sistema de tempo LTI de tempo cont√≠nuo e causal, cuja rela√ß√£o de entrada e sa√≠da √© descrita pela seguinte equa√ß√£o diferencial:
$$\ddot{y}(ùë°) + 11\dot{ùë¶}(ùë°) + 24ùë¶(ùë°) = 5\dot{ùë•}(ùë°) + 3ùë•(ùë°)$$

**(f)** **(M)** Simule esse sistema de tempo cont√≠nuo quando a entrada √© o degrau unit√°rio. Compare seus resultados com √†queles obtidos nos itens anteriores.

```python
import sympy as sympy
import control as control
sympy.init_printing()
t, s = sympy.symbols('t, s')

def Laplace(x):
    return sympy.laplace_transform(x, t, s, noconds=True)

def invLaplace(X):
    return sympy.inverse_laplace_transform(X, s, t)
```

Entrada:
$$x(t) = u(t)$$
$$\dot{u}(t) = \delta(t)$$
$$\mathcal{L}\{u(t)\} = \frac{1}{s}$$
$$\mathcal{L}\{\delta(t)\} = 1$$

Aplicando a entrada e, a transformada de Laplace em ambos lados da equa√ß√£o diferencial, temos:
$$
Y(s)[s^2 + 11s + 24] = 5 + \frac{3}{s}
$$

ou seja:

$$
Y(s) = \frac{5s + 3}{s(s^2 + 11s + 24)}
$$

```python
Y = (5*s + 3)/(s*(s**2 + 11*s +24))
Y
```

```python
y = invLaplace(Y)
y
```

$$\ddot{y}(ùë°) + 11\dot{ùë¶}(ùë°) + 24ùë¶(ùë°) = 5\dot{ùë•}(ùë°) + 3ùë•(ùë°)$$


$$H(s)=\frac{5s+3}{s^2+11s+24}$$

```python
import control as control
import numpy as np

num = np.array([5, 3])
den = np.array([1, 11, 24])
H = control.tf(num ,den)
H
```

```python
import matplotlib.pyplot as plt

t0 = 0
tend = 20
td = 1e-4
Npontos = int((tend - t0)/td)
T = np.linspace(t0,tend,Npontos)

t, h = control.step_response(H, T)

plt.plot(t, h)
plt.title('Resposta ao Degrau')
plt.xlabel("$t$")
plt.grid()
plt.show()
```
